```python
# Define a function to get the model's response
def get_model_response(prompt):
    try:
        response = client.chat.completions.create(
            model="ft:gpt-3.5-turbo-0125:personal::9cx0jt8Z",  # Your fine-tuned model
            messages=[
                {"role": "system", "content": "You are a helpful Amazon assistant and you have to resolve customer payment queries."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=150,
            n=1,
            stop=None,
            temperature=0.7
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error: {e}"

# Interactive chat loop
def chat_with_bot():
    print("You are now chatting with the Amazon Assistant. Type 'exit' to end the chat.")
    while True:
        user_input = input("User: ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        
        prompt = user_input
        response = get_model_response(prompt)
        print(f"Assistant: {response}")

# Start the chat
chat_with_bot()
```
